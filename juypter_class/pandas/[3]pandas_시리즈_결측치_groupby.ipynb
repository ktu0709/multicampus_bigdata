```python
#7) 결측치 확인 함수 
df = pd.read_csv('data/WHO_first9cols.csv')
df2 = df[['Country', df.columns[-2]]]
df2.info()
df3 = df2.fillna(0)
type(df3)
df3.isnull()
```

```python
# pd.isnull(), pd.notnull()
pd.isnull(df2)        # 결측치는 True로 반환, 나머지는 False로 반환
pd.notnull(df2)       # 결측치는 False로 반환, 나머지는 True로 반환
pd.isnull(df2).sum()  # 23개, -2번 컬럼의 결측치의 갯수
pd.isnull(df). sum()   # 전체 컬럼의 결측치의 갯수

# fillna() : 결측치를 특정 값으로 채움
# 결측치를 0으로 채움 :  평균, 표준 편차에 영향
df3 = df2.fillna(0)
pd.isnull(df3).sum()  # 0  , 결측치의 갯수는 0

df2.select_dtypes(include=[np.number]).mean()  # 85.698324
df3.select_dtypes(include=[np.number]).mean()  # 75.940594

# replace(a,b)
df4 = df2.replace(np.nan,0)  # fillna(0)와 동일한 결과

sr = df2[df2.columns[-1]].replace(np.nan,0)  # 한개의 컬럼만 결측치를 제거
sr

# 결측치를 평균으로 채움 : 평균에 영향이 없음
df5 = df2.fillna(df2.select_dtypes(include=[np.number]).mean())   # 
df5

# df2.mean()  # 85.698324
print(df2.describe())   # 결측치 제거전
print(df4.describe())   # 평균값이 감소
print(df5.describe())   # 평균값이 동일
```

### Series 객체의 index 사용 및 결측치 제어

```python
index_table = ['tv','vtr','phone']
sr = pd.Series([10,20,30],index=index_table)
print(sr)
print(type(sr))
print(sr.index)
r = sr.index.str.contains('v') # index에서 문자열을 검출
sr[r]
```

```python
# Series 객체의 요소의 추가
sr2 = pd.concat([sr , pd.Series([40,50,60],index=['audio','iron','notebook'])])
sr2
```

```python

sr3 = pd.concat([sr , pd.Series([np.nan],index=['aircon'])])
print(sr3)
print(sr3.mean())  # 35, NaN 을 제외하고 평균을 구함

# a = np.array([10,20,30,40,50,60,np.nan])
# a.mean()    # nan  , ndarray는 nan관 연산하면 결과값이 nan이다

# 결측치를 처리(제거)
sr4 = sr3.fillna(0)  # 0으로 채움
print(sr4)
print(sr4.mean())  # 30,  0으로 채우면 평균값이 감소

sr5 = sr3.fillna(sr3.mean())  # 평균값으로 채움
print(sr5)
print(sr5.mean())  # 35,  평균으로 채우면 평균값 변동이 없음
```

### groupby()

```python
# import numpy as np
# np.random.seed(55)
# np.random.rand(10,2)

from numpy.random import seed
from numpy.random import rand
from numpy.random import randint
seed(42)

df = pd.DataFrame({'Weather' : ['cold', 'hot', 'cold', 'hot', 'cold', 'hot', 'cold'],
                   'Food' : ['soup', 'soup', 'icecream', 'chocolate', 'icecream', 'icecream', 'soup'],
                   'Price' : 10 * rand(7),       # 0~10 사이,7개의 난수, 실수
                   'Number' : randint(1, 9, 7)}) # 1~9 사이,7개의 난수, 정수
df
```

```python
import pandas as pd
import numpy as np

weather_group = df.groupby('Weather')
weather_group
```

```python
import pandas as pd
import numpy as np

weather_group = df.groupby('Weather')

i = 0
for name, group in weather_group:
    numeric_group = group.select_dtypes(include=[np.number])
    i += 1
    print(f'[{i} {name}]\n{numeric_group}')
    print(type(numeric_group))  # DataFrame
    print('-' * 50)
    
```

```python
 
# 각 그룹의 첫번째 행
print(weather_group.first())
print('-' * 50)

# 각 그룹의 마지막 행
print(weather_group.last())
print('-' * 50)

# 각 그룹의 수치형 데이터에 대한 평균
print(numeric_group.mean())
print('-' * 50)

# 각 그룹의 수치형 데이터에 대한 총 갯수
print(numeric_group.count())
print('-' * 50)

# 각 그룹의 수치형 데이터에 대한 통계 요약
print(weather_group.describe())
```

```python
wf_group = df.groupby(['Weather','Food'])
print(wf_group)
print(wf_group.groups)  # dict, 행번호(value)

wf_group.agg([np.mean,np.sum,np.min,np.max,'max','mean'])
```

```python
gr_df = wf_group.agg([np.mean,np.sum,np.min,np.max])
print(type(gr_df))  # DataFrame (인덱스와 컬럼이 MultiIndex인 데이터 프레임
print(gr_df.index)  # MultiIndex
#print(gr_df)
print(gr_df.columns) # MultiIndex
```

```python
gr_df
```

```python
# MultiIndex형식의 컬럼 접근 방법-1
df1 = pd.DataFrame(gr_df['Price']['mean'])
df2 = pd.DataFrame(gr_df['Number']['mean'])
df3 = pd.concat([df1,df2],axis=1)
df3

# MultiIndex형식의 컬럼 접근 방법-2
gr_df.loc[:,[('Price', 'mean'),('Number', 'mean')]]
```
