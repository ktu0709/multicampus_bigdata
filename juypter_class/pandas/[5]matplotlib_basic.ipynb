선형 방정식 특징
1. 단일변수 : 단일 변수 X의 경우 선형방정식의 형식은 ax+b=0, a,b는 상수이다. ex) 3x+5=0
2. 두개이상의 변수 : 두개의 변수 x,y -> ax+by+c=0 a,b,c는 상수 , 2x-3y+6 = 0 -> 데카르트 평명위의 직선
3. 3개 이상의 변수 : x,y,z가 있는 선형방정식 -> ax+by+cz+d = 0  -> 3차원 공간의 그래프
4. 지수,복소수 연산 : 선형방정식에서 변수에는 1이외의 지수가 없으며 변수(xy),제곱근 또는 기타 결과가 없다 -> 방정식을 비선형으로 만들 때 사용  
5. 기울기, 절편 ; 2차원 선형방정식은 y=mx+b와 같은 기울기 - 절편 형식으로 작성될 수 있다. m은 선의 기울기, b는 y의 절편이다. -> 선이 교차하는 지점

선형방정식의 용도
1. 단일변수 : 재정 및 금융계산 (손익분기점, 비용분석) , 단일요인이 결과에 선형적으로 적용되는 예측
2. 두개이상의 변수 : 시간의 흐름을 분석할 때 , x= 시간 , y= 관심있는 변수
3. 3개 이상의 변수 : 다중회귀분석_2개 이상의 다른 변수의 값에 기초하여 변수의 값을 예측할 때
4. 지수,복소수 연산 : 최적화, 문제의 최적해를 찾을 때 (일정관리 , 자원할당)
5. 기울기, 절편 : 데이터 시각화 _ 두 변수 사이의 관계를 시각화 , 선형회귀에서 최적의 선을 그리기 위해 / 
                 예측분석 -> 데이터파생된 기울기 절편을 가지고 하나의 변수(x)의 값에 기반하여 다른 변수(y)에 예측값 

* 선형회귀 : 데이터 + 선형방정식 => 종속변수 하나(단순 선형회귀), 종속변수 여려개(다중 선형회귀)의 독립변수간의 관계를 모델링 한것
           
=> 선형 방정식의 계수는 데이터에서 추정되며 , 한개의 독립변수인 경우에는 데이터에 가장 잘맞는 선의기울기를 , 여러개의 독립변수인 경우는 가장 잘맞는 초평면(hyperplane)을 정의하는데 사용된다

초평면 : 차원을 -1한 후 고차원공간을 통과시키는 슬라이스이다

```python
# 1차 함수(방정식)의 그래프 : f(x) = a*x + b, 직선의 방정식
x = np.arange(-4.5,5.0,0.5)
y = 2*x + 0
print(x,y)

plt.plot(x,y)
plt.show()
```

```python
# 2차 함수(방정식)의 그래프 : f(x) = a*x^2 + b, 포물선의 방정식
x = np.arange(-4.5,5.0,0.5)
y = 2*x**2 + 0
print(x,y)

plt.plot(x,y)    # 포물선
plt.plot(x,4*x)  # 직선
plt.plot(x,-3*x) # 직선
plt.show()
```

### 로그 플롯(log plot)

```python
# 상용로그 : 10
print(np.log10(100))  # 2.0

# 자연로그 :  e = 2.71828
print(np.log(100))   # log(100) ==> 4.605170185988092
print(np.exp(4.605170185988092))  # e**4.605170185988092 ==> 100.0

#로그는 노이즈 제거할 때 사용된다
```

```python
df = pd.read_csv('data/transcount.csv')

# 연도별 CPU 판매량
df = df.groupby('year').agg(np.mean)
df
```

```python
years = df.index.values             # x값
counts = df['trans_count'].values   # y값

print(type(years),type(counts))     # ndarray
print(years,'\n',counts)

plt.plot(years,counts) # 원본 데이터로 그래프를 출력
                       # 올바른 정보(변화량)를 알 수가 없다
                       # 1970~2000년 사이의 변화량이 매우 작음
plt.show()
```

```python
# y값을 log변환하여 그래프로 출력
plt.plot(years,np.log(counts))   # 자연로그 사용 로그변환
                                 # 1970~2000년 사이의 변화량이 분명해짐
# np.log(counts)

# plt.plot(years,np.log10(counts))  # 상용 로그 
plt.show()
```

```python
print(np.log(1))   # 0.0
print(np.log(0))   # -inf

print(np.log1p(0)) # 0.0 , 1을 더한 후 log변환, log(0)이 -inf(무한대)가 되지 않도록
# np.log1p(x) ==> np.log(x + 1)
```

```python
# 로그 플롯 : semilogy(),semilogx(),loglog()
years = df.index.values             # x값
counts = df['trans_count'].values   # y값

plt.semilogy(years,counts,'o')  # semilogy함수는 y값을 로그변환하여 그래프로 출력

# 선형회귀 구현
poly = np.polyfit(years,np.log(counts),deg=1)
print(poly)  # [ 3.61559210e-01 -7.05783195e+02], 회귀 계수
# H(x) = 3.61559210e-01*x -7.05783195e+02 , 예측 함수

plt.semilogy(years,np.exp(np.polyval(poly,years))) # 예측함수의 직선
# plt.semilogy(years,np.exp(3.61559210e-01*years - 7.05783195e+02 ))
plt.show()

np.exp(np.polyval(poly,years)) # polyfit()으로 얻은 poly내에 기울기와
# y절편을 이용하여 예측함수에 x값(years)을 넣어 예측값(y,counts,H(x))을 구한다

# 2015년 판매량 예측
print(np.exp(3.61559210e-01*2015 -7.05783195e+02)) # 7654910314.116927
```

### 분산 플롯(scatter plot) : scatter()함수, 버블차트

```python
df = pd.read_csv('data/transcount.csv')
df = df.groupby('year').agg(np.mean)

gpu = pd.read_csv('data/gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df,gpu,how='outer',left_index=True,right_index=True)
df = df.fillna(0)     # 결측치를 0으로 수정
# df = df.replace(np.nan,0)
df
```

```python
years = df.index.values           # x값
counts = df['trans_count'].values 
cnt_log = np.log(counts)   # 로그변환한 y값
gpu_counts = df['gpu_trans_count'].values  # 버블의 크기

# plt.figure(figsize=(12,10))  # 그래프 사이즈의 크기 설정
plt.scatter(years,cnt_log, c = 200*years,
           s = 20 + 200*gpu_counts/gpu_counts.max(),
           alpha=0.5) # c : 컬러, s : 버블의 크기, alpah : 투명도
plt.grid(True)
plt.xlabel('years')
plt.ylabel('trans_count')
plt.title('trans & gpu counts per years')
plt.show()
```

```python
# 그래프에 한글 출력 설정(Windows)
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'   # '맑은 고딕'으로 설정 
matplotlib.rcParams['axes.unicode_minus'] = False      # minus 숫자 표시
matplotlib.rcParams['axes.titlesize'] = 18             # title 의 사이즈 변경
plt.title('연간판매량')
plt.show()
```

```python
# 범례(legend), 주석(annotation), 예측 함수의 직선 그래프를 함께 출력
years = df.index.values           # x값
counts = df['trans_count'].values 
cnt_log = np.log(counts)   # 로그변환한 y값
gpu_counts = df['gpu_trans_count'].values  # 버블의 크기

# 예측 함수의 직선 그래프를 출력
poly = np.polyfit(years,np.log(counts),deg=1)
plt.plot(years,np.polyval(poly,years),label='Poly Fit')

# annotaion(주석 달기)
gpu_start = gpu.index.values.min()        # x축 : 1997
y_ann = np.log(df.at[1997,'trans_count']) # y축 : 15.913528485217045
ann_str = 'First GPU\n%d'%gpu_start
# print(gpu_start)
# print(y_ann)

plt.annotate(ann_str,xy=(gpu_start,y_ann),
            arrowprops=dict(arrowstyle='->'),
            xytext=(-22,+30),textcoords='offset points')

# scatter plot
plt.scatter(years,cnt_log, c = 200*years,
           s = 20 + 200*gpu_counts/gpu_counts.max(),
           alpha=0.5,label='Scatter Plot') # c : 컬러, s : 버블의 크기, alpah : 투명도
plt.grid(True)
plt.xlabel('years')
plt.ylabel('trans_count')
plt.title('trans & gpu counts per years')
plt.legend(loc='upper left')    # 범례
plt.show()
```
