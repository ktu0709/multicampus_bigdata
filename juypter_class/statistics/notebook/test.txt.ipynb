1. 통계적 자료의 분류
통계적 자료는 양적(수치)과 질적(범주)로 나뉜다.
양적 자료: 수치로 표현되며 연속적인 값을 갖는다. 예시: 혈압, 온도.
질적 자료: 범주로 표현되며 명목 척도와 순서 척도로 나뉜다. 예시: 성별(명목), 선호도(순서).
2. 척도 수준
명목 척도: 범주 간에 순서가 없고 서열이 없다. 예시: 혈액형.
등간 척도: 범주 간에 순서가 있고 간격이 일정하다. 예시: 온도.
비율 척도: 절대적인 영점이 있고 비율이 의미가 있는 척도. 예시: 키, 무게.
서열 척도: 범주 간에 순서가 있지만 간격이 일정하지 않다. 예시: 만족도.
3. 자료의 대표적인 측도
평균: 양적 자료의 대표값으로 값의 합을 개수로 나눈 것.
중앙값: 양적 자료를 정렬했을 때 중앙에 위치한 값.
최빈값: 자료에서 가장 많이 나타나는 값.
4. 통계량과 분포의 특성
표준편차: 데이터의 분산 정도를 나타내는 측도로, 클수록 데이터가 흩어져 있다.
왜도: 분포의 비대칭 정도를 나타내며 양수면 오른쪽으로 치우쳐져 있고, 음수면 왼쪽으로 치우쳐져 있다.
5. 상관과 중심 경향성
상관 계수: 두 양적 변수 간의 선형 관계를 측정하는데 사용.
중심 경향성: 데이터의 중심을 나타내는 측도로, 평균과 중앙값이 대표적.
6. 시각화
히스토그램: 양적 자료의 분포를 시각적으로 표현하는 그래프.
상자 수염 그림(Boxplot): 데이터의 중앙값과 분포를 표현하는 그래프.
