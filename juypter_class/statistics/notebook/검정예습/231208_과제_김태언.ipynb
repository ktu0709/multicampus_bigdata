### 문제01] 새로운 약물의 효과에 대한 임상 시험 [귀무 가설을 공식화하고 테스트할 수 있는 실제 문제]

시나리오 :한 제약회사가 혈압을 낮추기 위한 새로운 약을 개발했습니다. 그들은 자신들의 약물이 현재의 표준 치료법보다 더 효과적이라고 주장합니다.
새로운 약물이 실제로 표준 치료법보다 더 효과적인지 테스트합니다. 

1)임상시험은 환자를 대상으로 진행됩니다. 환자는 무작위로 두 그룹으로 배정됩니다. 한 그룹은 새로운 약물을 받고 다른 그룹은 표준 치료를 받습니다.혈압 감소(mmHg 단위로 측정)는 고정된 치료 기간 이후 각 환자에 대해 기록됩니다

2)가설 수립를 생각해 봅니다.  

 - 귀무가설(H₀): 새로운 약물은 표준 치료법보다 더 큰 혈압 감소로 이어지지 않습니다.->  μnew ≤ μstandard로 표현
     -  μnew = 새로운 약물에 대한 평균 혈압 감소, μstandard= 표준 치료에 대한 평균
   
 - 대체 가설(H₁): 새로운 약물은 표준 치료법 또는 μnew > μstandard 보다 혈압을 더 크게 감소시킵니다.  

3)그룹의 환자 표본을 대상으로 혈압 강하 데이터를 수집했다고 가정합니다. 

4)검정 통계량 테스트
- 두 그룹의 평균을 비교하기 위해 독립적인 2-표본 t-검정(정규 분포 및 등분산 가정)을 수행할 가능성이 높습니다.
- 검정은 평균의 차이가 귀무가설을 지지하거나 반박할 수 있는 통계적으로 유의한지 여부를 결정합니다.

5)의사결정
- 테스트 결과, 새로운 약물의 평균 혈압 감소가 표준 치료법보다 훨씬 더 큰 것으로 나타나면(특정 신뢰도 이상, 일반적으로 95%) 귀무가설을 기각합니다.
- 테스트 결과가 통계적으로 유의하지 않으면 귀무 가설을 기각하지 못하며, 이는 새로운 약물이 더 효과적이라는 결론을 내리기에는 증거가 부족함을 나타냅니다.

```python
#1) 가설 데이터  :  두 그룹(새로운 약물로 치료한 그룹과 표준 치료로 치료한 그룹)에 대한 혈압 감소 데이터가 있다고 가정합니다. 
#신규 약물군: new_med = [10, 12, 14, 13, 11, 15, 14] (mmHg 단위 혈압 감소)
#표준치료군: standard_treat = [9, 11, 10, 8, 12, 9, 10]
```

```python
#2) 2-표본 T-검정 수행 
import numpy as np
from scipy import stats


new_med = [10, 12, 14, 13, 11, 15, 14]
standard_treat = [9, 11, 10, 8, 12, 9, 10]

#  t-test
t_statistic, p_value = stats.ttest_ind(new_med, standard_treat, equal_var=True)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

#### 통계적 해석
- T-통계는 그룹에서 관찰된 변동성을 기준으로 두 그룹 간의 차이를 측정
- P-값은  귀무가설이 참이라는 가정 하에 관찰된 결과만큼 극단적이거나 그보다 더 극단적인 결과를 관찰할 확률을 나타낸다 
- 결정 규칙은 일반적으로 p-값이 선택한 유의 수준(예: 5% 수준의 경우 0.05)보다 작은 경우 귀무 가설을 기각하고 새로운 약물과 표준 치료법 사이에 혈압 감소에 상당한 차이가 있음을 확인한다.

### 문제 2] 한 학교에서 학생의 수학 성적을 향상시키기 위해 새로운 교수법을 실험하고 있습니다.

학교에서는 이 새로운 방법이 기존 교육 방식보다 더 효과적인지 확인하려고 합니다.

1)목적 : 새로운 교수법이 기존 교수법에 비해 수학 평균 점수가 더 높은지 여부를 테스트합니다.

2)연구설계  
    - 학생들은 무작위로 두 그룹으로 배정됩니다. 
    - 한 그룹은 새로운 방법을 사용하여 가르치고 다른 그룹은 기존 방법을 계속 사용합니다.
    - 학기 말에는 모든 학생에게 표준화된 수학 시험이 실시됩니다.
    
3)가설 수립 
    - 귀무가설(H₀)
        -  새로운 교수법은 기존 방법에 비해 더 높은 평균 점수로 이어지지 않습니다. 
        -  통계적으로 이는 μnew ≤ μtraditional로 표현
        -  μnew: 새로운 방법의 평균 점수, μtraditional: 기존 방법의 평균
     - 대립 가설(H₁)
         - 새로운 교수법은 기존 방법 또는 μnew > μtraditional 
         
4)검정 통계
    - 독립적인 2-표본 t-검정을 실시하여 두 그룹의 평균 점수를 비교
    - 평균점수 차이가 통계적으로 유의미한지 판단
    
5)의사결정  
    - 테스트 결과에 따르면 새로운 방법의 평균 점수가 기존 방법보다 훨씬 높은 것으로 나타나면(특정 신뢰 수준 이상, 일반적으로 95%) 귀무 가설을 기각하게 됩니다.
   - 테스트 결과가 통계적으로 유의하지 않으면 귀무 가설을 기각하지 못하며, 이는 새로운 방법이 더 효과적이라는 결론을 내리기에는 증거가 부족함을 나타냅니다

```python
import numpy as np
from scipy import stats

np.random.seed(0)  
new_method_scores = np.random.normal(80, 5, 30)  # mean=80, std=5, 30 students
traditional_method_scores = np.random.normal(75, 5, 30)  # Traditional method, mean=75, std=5, 30 students

t_statistic, p_value = stats.ttest_ind(new_method_scores, traditional_method_scores, equal_var=True)

t_statistic, p_value  # T-통계: 6.6327 ,P-값: 약 1.21×10−8
```

#### 해석  
T-Statistic: 6.6327의 양수 값은 새로운 방법 그룹의 평균 점수가 기존 방법 그룹의 평균 점수보다 높음을 나타냅니다.
P-값: 매우 작은 p-값(0.05보다 훨씬 작음)은 귀무가설이 참인 경우(즉, 교육 방법에 차이가 없는 경우) 평균에서 이렇게 큰 차이를 관찰할 확률이 있음을 나타냅니다.  (매우 낮음)
#### 결론
낮은 p-값이 주어지면 귀무 가설을 기각합니다. 이는 표본 데이터를 토대로 새로운 교수법이 기존 교수법에 비해 수학 평균 점수가 더 높다는 결론을 내릴 수 있는 통계적으로 유의미한 증거가 있음을 시사

### 문제 3] 새로운 마케팅 캠페인의 영향

- 최근 한 회사가 매장의 일일 평균 매출을 늘리기 위해 새로운 마케팅 캠페인을 시작했습니다. 그들은 이 캠페인의 효과를 평가하고 싶어합니다.

- 목적: 새로운 마케팅 캠페인으로 인해 일일 평균 매출이 크게 증가했는지 확인합니다.

1)연구 설계
    - 캠페인 전 기간: 캠페인 시작 전 기간 동안의 일평균 매출 데이터를 수집합니다.  
    - 캠페인 종료 후 기간: 충분한 시간 동안 캠페인을 진행한 후, 새로운 일일 평균 매출 데이터를 수집합니다.
    
2)가설 수립  
     
     - 귀무가설(H₀): 마케팅 캠페인은 일일 평균 매출에 영향을 미치지 않습니다.   
        μpost≤ μpre 로 표현, μpost= 캠페인 이후의 평균 매출, μpre = 캠페인 전의 평균 매출
        
     - 대립 가설(H₁): 마케팅 캠페인으로 인해 일일 평균 매출, 즉 μpost> μpre가 증가했습니다.
    
3)통계 테스트 
     - 캠페인 전후의 평균 매출을 비교하기 위해 독립적인 2표본 t-검정(정규분포 및 등분산 가정)을 실시합니다.
     - 평균 매출의 차이가 통계적으로 유의한지 여부를 확인합니다.
     
4)테스트 결과에서 캠페인 후 평균 매출이 크게 증가한 것으로 나타나면(p-값 < 선택한 유의 수준, 일반적으로 0.05) 귀무 가설을 기각합니다.


```python
#통계 테스트
import numpy as np
from scipy import stats

np.random.seed(0)  
pre_campaign_sales = np.random.normal(500, 20, 30)  # Pre-campaign, mean=500, std=20
post_campaign_sales = np.random.normal(530, 20, 30)  # Post-campaign, mean=530, std=20

t_statistic, p_value = stats.ttest_ind(post_campaign_sales, pre_campaign_sales)

t_statistic, p_value  
#T-통계: 2.938977037944546
#P-값: 0.004720335438244019
```

#### 해석
- T-통계: 양수 값은 캠페인 후 평균 매출이 캠페인 전 매출보다 높다는 것을 나타냅니다.
- P-값: p-값은 대략 0.0047로, 일반적인 유의 수준인 0.05보다 낮습니다.
#### 결론
통계 테스트에서는 p-값이 0.0047이고 양의 t-통계량을 사용하여 마케팅 캠페인이 효과적으로 매출을 늘린 것으로 나타났습니다.



### 문제4]   다이어트약_효과검증.csv로 [문제3]번 형식으로 만들어 제출하세요 

- 최근 한 회사가 다이어트 약을 새로 개발하여 임상테스트를 진행했습니다. 테스트 결과 데이터를 기반으로 다이어트 약의 효과를 입증하고싶습니다

- 목적: 다이어트약이 체중감소에 효과를 준다는것을 확인하고 싶습니다

1)연구 설계
    - 다이어트전(kg): 다이어트약 복용전 실험자들의 체중 데이터를 수집합니다.  
    - 다이어트후(kg): 다이어트약 복용후 실험자들의 체중 데이터를 수집합니다.
        
2)가설 수립       
     - 귀무가설(H₀): 다이어트약이 체중감소에 영향을 주지 않는다    
        μafter <= μbefore 로 표현, μafter= 다이어트후(kg), μbefore = 다이어트전(kg)
        
     - 대립 가설(H₁): 다이어트약이 체중감소에 영향을 준다, 즉 μafter > μbefore로 표현      
        
3)통계 테스트   
     - 다이어트약 복용 전/후 체중을 비교하기 위해 대응 표본 t-검정을 실시합니다.   
     - 다이어트약 복용 전/후 체중이 통계적으로 유의한지 여부를 확인합니다.
     
4)테스트 결과에서 테스트 결과에서 다이어트약이 체중감소를 나타나면(p-값 < 선택한 유의 수준, 일반적으로 0.05) 귀무 가설을 기각합니다.

```python
import pandas  as pd
from scipy import stats

df_4 = pd.read_csv('./data/다이어트약_효과검증.csv')
df_4_B = df_4['다이어트전(kg)']
df_4_A = df_4['다이어트후(kg)']

t_statistic, p_value = stats.ttest_rel(df_4_B, df_4_A)

t_statistic, p_value
#T-통계: 9.70596986565785
#P-값: 5.370404098342324e-13
```

해석
- T-통계: 양수 값은 다이어트약이 체중감소에 영향을 준다것으로 나타냅니다.
- P-값: p-값은 대략 5.370404098342324e-13로, 일반적인 유의 수준인 0.05보다 낮습니다.

결론
통계 테스트에서는 p-값이 5.370404098342324e-13이고 양의 t-통계량을 사용하여 다이어트약이 체중감소에 영향을 주는 것으로 나타났습니다.

### 문제5]  고객 만족을 위한 두 가지 커피 추출 방법 평가 [나머지 마무리 해서 제출하세요]

현지 인기 커피숍 '카페인 허브'는 완벽한 한 잔의 커피를 만들기 위해 노력하는 곳으로 알려져 있습니다. 경영진은 고객 만족도를 높이기 위해 다양한 양조 방법을 실험해 왔습니다. 최근에는 전통적인 '에스프레소' 추출 기술과 함께 새로운 '슬로우 드립(Slow Drip)' 방식을 도입했습니다.


1) 목적 : 새로운 슬로우 드립 방식이 기존 에스프레소 방식에 비해 고객 만족도를 높이는지 확인합니다.
2) 연구설계 :  
         - 전통적인 에스프레소 추출 기술만든 커피의 만족도  
         - '슬로우 드립(Slow Drip)' 방식으로 내린 커피의 만족도

- 참가자: 100명의 일반 고객이 연구에 참여하도록 무작위로 선택됩니다.
- 개입: 고객을 무작위로 50명씩 두 그룹으로 나눕니다. 한 그룹에는 슬로우 드립 방식으로 추출한 커피가 제공되고, 다른 그룹에는 전통적인 에스프레소 방식으로 추출된 커피가 제공됩니다.
- 체험: 참가자들은 매장 내 편안한 분위기에서 커피를 즐깁니다.
- 피드백 수집: 커피를 마신 후 각 참가자는 표준화된 만족도 설문조사를 완료하여 자신의 경험을 0(매우 불만족)부터 100(매우 만족)까지 점수를 매깁니다.

3) 가설수립:
     - 귀무가설(H₀): 두 커피의 평균 만족도가 비슷하다.   
        μnew≤ μespresso 로 표현, μnew= 슬로우 드립 커피 평균 만족도, μespresso = 에스프레소 커피 평균 만족도
        
     - 대립 가설(H₁):두 커피의 평균 만족도가 비슷하지 않다, 즉 μnew >= μespresso 
     
3)통계 테스트   
     - 캠페인 전후의 평균 매출을 비교하기 위해 독립적인 2표본 t-검정(정규분포 및 등분산 가정)을 실시합니다.  
     - 두 커피의 평균 만족도에 차이가 있는지 확인합니다     

```python
import numpy as np
from scipy import stats

np.random.seed(0)  
df_5_A = np.random.normal(50, 20, 50)
df_5_B = np.random.normal(55, 20, 50)

t_statistic, p_value = stats.ttest_ind(df_5_A, df_5_B)

t_statistic, p_value
#T-통계: -0.43601615616363504
#P-값: 0.6637839805373993
```

해석
T-통계: 음수 값은 두 커피의 평균 만족도가 비슷하다는 것을 나타냅니다.
P-값: p-값은 대략 0.663로, 일반적인 유의 수준인 0.05보다 높습니다.
    
결론 통계 테스트에서는 p-값이 0.663이고 음의 t-통계량을 사용하여  두 커피의 평균 만족도가 비슷하다는 결론이 나타났습니다.
